```python
import numpy as np
import pandas as pd
from scratch import *
```

```python
preprocess = MyPreProcessor()
X,Y=preprocess.pre_process(2)
print(X)
print(Y)
```

```python
ratio = X.shape[0]//10
X_train = X[0:7*ratio, :]
y_train = Y[0:7*ratio, :]
X_val = X[7*ratio : 8*ratio, :]
y_val = Y[7*ratio : 8*ratio, :]
X_test = X[8*ratio:, :]
y_test = Y[8*ratio:,:]
print(X_train.shape,y_train.shape )
print(X_test.shape,y_test.shape )
print(X_val.shape, y_val.shape)
```

```python
pp = MyPreProcessor()
W = pp.get_analytical_sol()
print(W)
X,y = pp.pre_process(0)
bias = np.zeros((X.shape[0],1))
bias.fill(1)
X = np.append(X,bias, axis = 1)
y_pred = np.dot(X,W)
error = (np.sum((y-y_pred)**2)/y_pred.shape[0])**0.5
mae = np.sum(abs(y-y_pred))/y_pred.shape[0]
print(error)
print(mae)
```

```python
"""Running SGD on Banknote dataset and with LR = 0.0001, 0.01, 10"""
L = [0.0001,0.01,10]
```

```python
for i in range(3):
    print("\n For learning rate ", L[i], "------------------------------------------------------------ ")
    log_regressor_sgd = MyLogisticRegression()
    log_regressor_sgd.fit(X_train, y_train, X_test, y_test, epochs = 150 ,learning_rate = L[i], grad_type = "sgd")
    print(log_regressor_sgd.W)
    print(log_regressor_sgd.b)
    loss_history_sgd = log_regressor_sgd.loss_history
    val_loss_history_sgd = log_regressor_sgd.val_loss_history
    train_acc_history_sgd = log_regressor_sgd.train_acc_history
    val_acc_history_sgd = log_regressor_sgd.val_acc_history
    plt.plot([x for x in range(151)], loss_history_sgd, label = "Training  loss " )
    plt.plot([x for x in range(151)], val_loss_history_sgd, label = "Validation  loss")

    plt.legend()
    plt.show()
    plt.plot([x for x in range(151)], train_acc_history_sgd, label = "Training  accuracy " )
    plt.plot([x for x in range(151)], val_acc_history_sgd, label = "Validation  accuracy")

    plt.legend()
    plt.show()
    
    print("\n -------------------------------------------------------------------------------------------")
```
